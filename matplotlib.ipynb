```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
x = [1,2,3,4,5]
x=np.array(x)
y=x**2
plt.title('X^2')
plt.xlabel('1-5')
plt.ylabel('1-5 squared')
plt.plot(x,y)
plt.show()
```

```python
#2 สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู
x=np.arange(0,90)
y=np.sin(x)
plt.title('Sin graph')
plt.xlabel('0-90')
plt.ylabel('-1-1')
plt.plot(x,y,c='m')
plt.show()
```

```python
#3 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
#เส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
y=np.arange(0,10)
x=y**2
plt.title('X=Y^2')
plt.xlabel('y square')
plt.ylabel('0-10')
plt.plot(y,x,c='#ADFF2F')
plt.show()
```

```python
#4สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
#เส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -

x=np.arange(0,90)
y=np.cos(x)
plt.title('Cos graph')
plt.xlabel('0-90')
plt.ylabel('cos x ')
plt.plot(x,y,c='r',ls='--')
plt.show()
```

```python
#5. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อ
#กราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker
#และเส้นสองใช้สามเหลี่ยมเป็น marker

x=np.arange(0,10)
plt.plot(x,x**2,label='Y=X^2 function',c='r',marker='o')
plt.plot(x,x**3,label='Y=X^3 function',c='b',marker='^')
```

```python
#7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
#เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม

y=np.arange(0,10)
x=(y**2)+(4*y)
plt.plot(y,x,c='#FFFACD')
plt.xlim([0,10])
plt.ylim([0,150])
plt.show()
```

```python
#8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน
#พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 


y1=np.arange(0,10)
x1=y1**3
x2=np.arange(0,10)
y2=(x2**2)+2
fig=plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.9,0.8])
ax1.plot(x1,y1,label = 'X = Y^3')
ax1.legend(loc = 0)
ax1.set_xlabel('y cubed')
ax1.set_ylabel('0-10')
ax1.set_title('X = Y^3')

ax2 = fig.add_axes([0.5 ,0.2, 0.4 , 0.4])
ax2.plot(x2,y2,label = 'Y = X^2 + 2 ')
ax2.legend(loc = 0)
ax2.set_xlabel('0-10')
ax2.set_ylabel('x^2+2')
ax2.set_title('Y = X^2 + 2')
```
